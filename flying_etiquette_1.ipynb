### 2. How did survey respondents generally feel regarding if “it [was] rude to ask someone to switch seats with you in order to be closer to friends?”?

```python
column_friends ='Is it rude to ask someone to switch seats with you in order to be closer to friends?'
flying[column_friends].head()
```

```python
flying[column_friends].value_counts(dropna=False)
```

```python
f = flying.groupby(column_friends)
```

```python
f.groups
```

```python
f.size()
```

```python
f.size().plot(kind='bar', rot = 45, ylabel="Count", title=column_friends, xlabel='Proximity to Friends')
```

```python
f.size().plot(kind='barh', xlabel="Count", title=column_friends, ylabel='Proximity to Friends')
```

### 3. Use the respondents age as a dimension to see if there are any differences to responses to “it [was] rude to ask someone to switch seats with you in order to be closer to friends?”

```python
sns.countplot(flying, x=column_friends, hue="Age")
```

```python
sns.countplot(flying, x=column_friends, hue="Gender")
```

### Multiple plots

```python
# Two side-by-side plots:
plt.figure()
ax1 = plt.subplot(1,2,1)
# first argument is number of rows, second is number of columns, third is current plot number
sns.countplot(flying, x=column_friends, hue="Age")
ax2 = plt.subplot(1,2,2, sharey=ax1) 
sns.countplot(flying, x=column_friends, hue="Gender")
```

```python
# Two top and bottom charts:
plt.figure()
# first argument is number of rows, second is number of columns, third is current plot number
ax1 = plt.subplot(2,1,1)
sns.countplot(flying, x=column_friends, hue="Age")
ax2 = plt.subplot(2,1,2, sharey=ax1) 
sns.countplot(flying, x=column_friends, hue="Gender")
```

```python
sns.histplot(flying, x = column_friends, hue= "Gender", multiple = 'stack')
```
